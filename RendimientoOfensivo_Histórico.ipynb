```python
df = pd.read_csv('/content/drive/MyDrive/TFM/datasetSeleccionadasPreClustering.csv', decimal='.')
```

```python
#Eliminacion promedios y columnas que no se van a utilizar
eliminacion_clumnas_menor_60 = ['id', 'home_clearances', 'home_fouls_conceded', 'home_offsides', 'home_tackles', 'away_clearances', 'away_fouls_conceded',
                                'away_offsides', 'away_tackles', 'clearances_avg_H', 'fouls_conceded_avg_H', 'offsides_avg_H', 'tackles_avg_H',
                                'fouls_conceded_avg_A', 'offsides_avg_A', 'tackles_avg_A']
df.drop(eliminacion_clumnas_menor_60, axis=1, inplace=True)
df
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
import matplotlib.pyplot as plt

# Características ofensivas relevantes
offensive_features_home = [
    'home_shots', 'home_shots_on_target', 'goal_home_ft'
]

offensive_features_away = [
    'away_shots', 'away_shots_on_target', 'goal_away_ft'
]

df_away = df[offensive_features_away + ['away_team', 'season']].copy()
df_away.columns = ['shots', 'shots_on_target', 'goals', 'team', 'season']

df_home = df[offensive_features_home + ['home_team', 'season']].copy()
df_home.columns = ['shots', 'shots_on_target', 'goals', 'team', 'season']

# Combinar estadísticas de home y away
df_combined = pd.concat([df_home, df_away])

# Calcular estadísticas ofensivas
team_stats = df_combined.groupby('team').agg({
    'shots': 'mean',
    'shots_on_target': 'mean',
    'goals': 'mean'
}).reset_index()

# Calcular el número de temporadas jugadas por cada equipo
seasons_played = df_combined.groupby('team')['season'].nunique().reset_index()
seasons_played.columns = ['team', 'seasons_played']

team_stats = team_stats.merge(seasons_played, on='team', how='left')

scaler = StandardScaler()
data_scaled = scaler.fit_transform(team_stats.drop(columns=['team', 'seasons_played']))

# Método Elbow
sse = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_scaled)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k_range, sse, marker='o')
plt.xlabel('Número de clusters')
plt.ylabel('SSE')
plt.title('Método del Codo')
plt.show()

#K-means con el número de clusters deseado (4)
kmeans = KMeans(n_clusters=4, random_state=42)
team_stats['cluster'] = kmeans.fit_predict(data_scaled)

# Evaluación de la calidad del clustering
silhouette_avg = silhouette_score(data_scaled, team_stats['cluster'])
davies_bouldin_avg = davies_bouldin_score(data_scaled, team_stats['cluster'])
calinski_harabasz_avg = calinski_harabasz_score(data_scaled, team_stats['cluster'])

print(f'Silhouette Score: {silhouette_avg}')
print(f'Davies-Bouldin Index: {davies_bouldin_avg}')
print(f'Calinski-Harabasz Index: {calinski_harabasz_avg}')

# Asignar categorías de rendimiento ofensivo a cada cluster
category_map = {
    0: 'Muy Alta Capacidad Ofensiva',
    1: 'Alta Capacidad Ofensiva',
    2: 'Baja Capacidad Ofensiva',
    3: 'Muy Baja Capacidad Ofensiva'
}
team_stats['offensive_category'] = team_stats['cluster'].map(category_map)

category_order = ['Muy Alta Capacidad Ofensiva', 'Alta Capacidad Ofensiva', 'Baja Capacidad Ofensiva', 'Muy Baja Capacidad Ofensiva']
team_stats['offensive_category'] = pd.Categorical(team_stats['offensive_category'], categories=category_order, ordered=True)
team_stats = team_stats.sort_values('offensive_category')

output_file_path = '/content/drive/MyDrive/TFM/ofensive_clusters_modified.csv'
team_stats.to_csv(output_file_path, index=False)
print(f'Resultados guardados en {output_file_path}')

# Clasificación de los equipos
team_classification = team_stats[['team', 'shots', 'shots_on_target', 'goals', 'seasons_played', 'offensive_category']]
print(team_classification.head(10))

plt.figure(figsize=(12, 8))
scatter = plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=team_stats['cluster'], cmap='viridis')
plt.title('Visualización de Clusters de Rendimiento Ofensivo')
plt.xlabel('Característica Normalizada 1')
plt.ylabel('Característica Normalizada 2')
plt.legend(handles=scatter.legend_elements()[0], labels=category_order, title='Capacidad Ofensiva')
plt.show()
```
